### Standardization (or Z-score Normalization)
Standardization is the process of rescaling the values of features so that they have a mean of 0 and a standard deviation of 1. This technique is often used in data preprocessing to ensure that features with different scales don‚Äôt disproportionately affect the performance of machine learning models. The formula for standardization is known as the Z-score formula: ùëç = (ùëã ‚àí ùúá) / ùúé
<br> Where:
<br> ùëç is the ùëç-score or standardized value.
<br> ùëã is the original value of the feature.
<br> ùúá is the mean of the feature.
<br> ùúé is the standard deviation of the feature.

```python
import seaborn as sns
import numpy as np

df = sns.load_dataset('tips')
df
```

```python
bill = list(df['total_bill'])
mean = np.mean(bill)
std = np.std(bill)

normalized_bill = []

for i in bill :
    z_score = (i - mean) / std
    normalized_bill.append(z_score)

print(normalized_bill)
```

```python
normalized_mean = np.mean(normalized_bill)
print('Mean after Normalization: ', normalized_mean)
np.isclose(normalized_mean, 0)
```

```python
# applies standardization to the total_bill and tip columns in your df DataFrame using the StandardScaler from the sklearn.preprocessing module

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # creates an instance of the StandardScaler
scaler
```

^This line imports StandardScaler, which is a preprocessing tool used to standardize features in a dataset. It scales the data so that each feature has a mean of 0 and a standard deviation of 1.

```python
scaler.fit(df[['total_bill', 'tip']]) 
# The fit() method computes the mean and standard deviation for each feature (i.e., total_bill and tip), which the scaler will use to transform the data.
```

```python
scaler.transform(df[['total_bill', 'tip']])
# After fitting, you can use the scaler to standardize the data
```

```python
import pandas as pd

pd.DataFrame(scaler.transform(df[['total_bill', 'tip']]), columns = ['Total Bill', 'Tips'])
```
